```python
def easy_data(start_date, end_date, eng):
    """ total_easy_data('2022-01-01', '2022-06-30', local_engine)  """
    
    # BuySell 데이터 처리
    buysell = pd.read_sql(f"SELECT * FROM buysell WHERE adate >= #{start_date}# AND adate <= #{end_date}#", eng)
    buysell['adate'] = pd.to_datetime(buysell['adate']).dt.strftime('%Y-%m-%d')
    buysell['year'] = buysell['adate'].str[:4]
    buysell['month'] = buysell['adate'].str[5:7]
    buysell = buysell[['adate', 'atype', 'o_num', 'year', 'month', 'total_price']]
    buysell.columns = ['adate', 'atype', 'num', 'year', 'month', 'total_price']

    # PayIncome 데이터 처리
    payincome = pd.read_sql(f"SELECT * FROM call WHERE cdate >= #{start_date}# AND cdate <= #{end_date}#", eng)
    payincome['cdate'] = pd.to_datetime(payincome['cdate']).dt.strftime('%Y-%m-%d')
    payincome['year'] = payincome['cdate'].str[:4]
    payincome['month'] = payincome['cdate'].str[5:7]
    payincome = payincome[['cdate', 'atype', 'o_num', 'year', 'month', 'call_price']]
    payincome.columns = ['cdate', 'atype', 'num', 'year', 'month', 'call_price']

    # 표준 사무실 데이터 로드 및 병합
    standard_office = pd.read_excel(f"{path_balance_standard}standard_office.xlsx")
    buysell = pd.merge(buysell, standard_office, on='num', how='left').dropna()
    payincome = pd.merge(payincome, standard_office, on='num', how='left').dropna()

    # 데이터 정렬
    buysell = buysell.sort_values(['year', 'month', 'onames'], ascending=[True, True, True])
    payincome = payincome.sort_values(['year', 'month', 'onames'], ascending=[True, True, True])

    # Buy, Sell, Pay, Income 데이터 분리
    buy = buysell[buysell.atype == 1][['year', 'month', 'num', 'onames', 'total_price']]
    sell = buysell[buysell.atype == 2][['year', 'month', 'num', 'onames', 'total_price']]
    pay = payincome[payincome.atype == 4][['year', 'month', 'num', 'onames', 'call_price']]
    income = payincome[(payincome.atype == 2) | (payincome.atype == 3)][['year', 'month', 'num', 'onames', 'call_price']]

    # 그룹화 및 합계 계산
    def group_and_sum(df, value_column, new_column_name):
        grouped = df.groupby(['year', 'month', 'num', 'onames'])[value_column].sum().reset_index()
        grouped = grouped.rename(columns={value_column: new_column_name})
        return grouped.sort_values(by=['year', 'onames']).reset_index(drop=True)

    month_buy_sum = group_and_sum(buy, 'total_price', 'easy_buy')
    month_sell_sum = group_and_sum(sell, 'total_price', 'easy_sell')
    month_pay_sum = group_and_sum(pay, 'call_price', 'easy_pay')
    month_income_sum = group_and_sum(income, 'call_price', 'easy_income')

    """# 엑셀 파일 저장
    with pd.ExcelWriter(f'{path_balance_output}total_easy_data.xlsx', engine='xlsxwriter') as writer:
        month_buy_sum.to_excel(writer, sheet_name='buy', index=False)
        month_sell_sum.to_excel(writer, sheet_name='sell', index=False)
        month_pay_sum.to_excel(writer, sheet_name='pay', index=False)
        month_income_sum.to_excel(writer, sheet_name='income', index=False)"""
    
    return month_buy_sum, month_sell_sum, month_pay_sum, month_income_sum
```

```python
def bank_data(start_year, end_year):
    """total_bank_data(2019, 2024)"""
    st = pd.read_excel(f"{path_balance_standard}standard_office.xlsx")
    # 단순 치환 리스트
    simple_change = ['KT', '알바', '경동택배', '식비', '식대', '주민세', '주차', '급여', '윈텍', '현대자동차', '지엔피테크', '조은전자', 
                     '전진테크', '유진전자', '전기요금', '자동차세', '임대료', '에니컴포넌트']
    
    total_year_coltd_bank = pd.DataFrame()
    total_year_vip_bank = pd.DataFrame()
    total_year_promise = pd.DataFrame()
    
    # start_year부터 end_year까지의 범위 설정
    which_year = [str(year) for year in range(start_year, end_year + 1)]
    
    for i in which_year:
        try:
            promise = pd.read_excel(f"{path_balance_sheet}promise_{i}.xlsx")
            total_year_promise = pd.concat([total_year_promise, promise], ignore_index=True)
        except Exception as e:
            pass  # Skip if there's any issue with the file      
        
        coltd_bank = pd.read_excel(f"{path_balance_sheet}coltd_bank_{i}.xlsx", skiprows=9, index_col=0)
        vip_bank = pd.read_excel(f"{path_balance_sheet}vip_bank_{i}.xlsx", skiprows=7, skipfooter=1, index_col=1)
        total_year_coltd_bank = pd.concat([total_year_coltd_bank, coltd_bank], ignore_index=True)
        total_year_vip_bank = pd.concat([total_year_vip_bank, vip_bank], ignore_index=True)     
        
        
    total_year_promise = total_year_promise[['거래일자', '어음번호', '어음금액']]
    total_year_promise.columns = ['date', 'num', 'income']

    # Convert 'date' and 'num' to string, get the first digit of 'num' and perform replacements
    total_year_promise['date'] = total_year_promise['date'].astype(str)
    total_year_promise['num'] = total_year_promise['num'].astype(str).str[:1].astype(int).replace({4: 257, 2: 304})

    # Clean and convert 'income' to numeric
    total_year_promise['income'] = total_year_promise['income'].str.replace(' 원', '').str.replace(',', '').astype(int)

    # Extract 'year' and 'month' from 'date'
    total_year_promise['year'] = total_year_promise['date'].str[:4]
    total_year_promise['month'] = total_year_promise['date'].str[4:6]
    
    # Extract 'year' and 'month' from 'date'
    total_year_promise['year'] = total_year_promise['date'].str[:4].astype(int)
    total_year_promise['month'] = total_year_promise['date'].str[4:6].astype(int)
    total_year_promise =total_year_promise[['year', 'month', 'num', 'income']]
    total_year_promise = pd.merge(total_year_promise, st, on='num', how='left')
    #total_year_promise['income'] = total_year_promise['income'].str.replace(' 원', '').str.replace(',', '').astype(int)
    # Extract 'year' and 'month' from 'date'
    #total_year_promise['year'] = total_year_promise['date'].str[:4]
    #total_year_promise['month'] = total_year_promise['date'].str[4:6].astype(int)
    total_year_promise =total_year_promise[['year', 'month', 'num', 'income']]
    total_year_promise = pd.merge(total_year_promise, st, on='num', how='left')
    
    # 컬럼 이름 및 데이터 처리
    total_year_coltd_bank = total_year_coltd_bank[['거래일자', '출금금액(원)', '입금금액(원)', '거래내용', '거래기록사항']]
    total_year_coltd_bank.columns = ['date', 'pay', 'income', 'message', 'onames']
    total_year_coltd_bank['year'] = total_year_coltd_bank['date'].str.slice(0, 4).astype(int)
    total_year_coltd_bank['month'] = total_year_coltd_bank['date'].str.slice(5, 7).astype(int)
    total_year_coltd_bank['onames'] = total_year_coltd_bank['onames'].fillna(total_year_coltd_bank['message'])  # NaN일 때 message로 대체
    total_year_coltd_bank = total_year_coltd_bank.reset_index(drop=True)

    # 단순 치환 작업
    for i in simple_change:
        total_year_coltd_bank['onames'] = total_year_coltd_bank['onames'].str.replace(rf'(^.*{i}.*$)', i, regex=True)
    
    # 문자열 치환 작업
    replacements = [
        (r'\d+/.*$', '대체'),
        (r'.*연호.*$', '연호전자'),
        (r'.*ENG.*$', '조인ENG'),
        (r'.*아성사.*$', '아성사'),
        (r'.*세종전자.*$', '세종전자'),
        (r'.*올넷.*$', '올넷전자'),
        (r'원텍', '윈텍'),
        (r'.*에바다$', '에바다전기'),
        (r'^김태원.*$', '에스이'),
        (r'.*에스미$', '에스이'),
        (r'^해인전자.*$', '해인전자'),
        (r'^조양산업.*$', '조양산업'),
        (r'^전병군.*$', '광진하네스'),
        (r'^서진전자.*$', '서진전자'),
        (r'^엄창주.*$', '제이투기전'),
        (r'^손남익.*$', '동원테크'),
        (r'^동원테크.*$', '동원테크'),
        (r'^김종학\(남웅.*$', '남웅전자'),
        (r'^송성용\(엑셀.*$', '엑셀전자'),
        (r'.*한솔하이테크.*$', '한솔하이테크'),
        (r'.*대흥전자$', '대흥전자부품'),
        (r'^서명섭.*$', '마이크로하네스'),
        (r'^제원.*$', '제원테크'),
        (r'^환경개선부.*$', '환경개선부담금'),
        (r'^합산보험.*$', '합산보험료'),
        (r'^보험료.*$', '차량보험료'),
        (r'.*민재성회계$', '민재성세무사'),
        (r'^[1-9].*보험료.*$', '차량보험료'),
        (r'^이도중.*$', '이도중'),
        (r'^.*식대.*$', '식비'),
        (r'.*경찰서$', '범칙금'),
        (r'^과태료.*$', '범칙금'),
        (r'^퇴직수수료$', '퇴직연금수수료'),
        (r'^퇴직연금$', '퇴직연금수수료'),
        (r'.*곽효빈.*$', '알바'),
    ]

    for pattern, replacement in replacements:
        total_year_coltd_bank['onames'] = total_year_coltd_bank['onames'].str.replace(pattern, replacement, regex=True)

    # VIP bank 데이터 처리
    total_year_vip_bank = total_year_vip_bank.dropna(axis=1, how='all')
    total_year_vip_bank = total_year_vip_bank[['거래일시', '출금금액', '입금금액', '거래내용', '거래기록사항']]
    total_year_vip_bank.columns = ['date', 'pay', 'income', 'message', 'onames']
    total_year_vip_bank['year'] = total_year_vip_bank['date'].str.slice(0, 4).astype(int)
    total_year_vip_bank['month'] = total_year_vip_bank['date'].str.slice(5, 7).astype(int)
    total_year_vip_bank['onames'] = total_year_vip_bank['onames'].fillna(total_year_vip_bank['message'])  # NaN일 때 message로 대체
    total_year_vip_bank = total_year_vip_bank.reset_index(drop=True)
    
    bank = pd.concat([total_year_coltd_bank, total_year_vip_bank])
    df2 = pd.read_excel(f"{path_balance_standard}simple_bank_office.xlsx")
    # 'onames' 열을 비교해  grouped_bank_income의 onames에 df2의 onames가 포함되어 있는지 확인하고 매핑
    bank['num'] =  bank['onames'].apply(lambda x: df2[df2['onames'].apply(lambda y: y in x)]['num'].values[0] if df2['onames'].apply(lambda y: y in x).any() else None)    
    bank = bank[['year', 'month', 'num', 'onames', 'pay', 'income']]
    bank = pd.concat([bank, total_year_promise]).reset_index()
    bank = pd.merge(bank, st, on='num' , how='left')
    
    bank = bank[['year', 'month', 'num', 'onames_y', 'pay', 'income']]
    bank.columns = ['year', 'month', 'num', 'onames', 'pay', 'income']
    
    bank_pay = bank[['year', 'month', 'num', 'onames', 'pay']]
    bank_income = bank[['year', 'month', 'num', 'onames', 'income']]
    
    bank_pay = bank_pay.dropna()
    bank_income = bank_income.dropna()
    
    bank_pay= bank_pay.groupby(['year', 'month', 'num', 'onames'])['pay'].sum().reset_index()
    bank_income = bank_income.groupby(['year', 'month', 'num', 'onames'])['income'].sum().reset_index()
    
    """#최종 데이터를 Excel 파일로 저장
    with pd.ExcelWriter(f"{path_balance_output}total_bank_data.xlsx", engine='xlsxwriter') as writer:
        bank_pay.to_excel(writer, sheet_name='pay', index=False)
        bank_income.to_excel(writer, sheet_name='income', index=False)"""
    
    return bank_pay, bank_income  
```
